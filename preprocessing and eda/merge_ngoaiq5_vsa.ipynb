```python
import pandas as pd
import os

# Đường dẫn đến 3 file cần merge
file_ngoaiq5 = '../dataset/preprocessed/reviews_final_ngoaiq5.csv'
file_vsa = '../dataset/preprocessed/reviews_final_vsa.csv'
file_vsa_test = '../dataset/preprocessed/reviews_final_vsa_test.csv'

# Đọc 3 file
df_ngoaiq5 = pd.read_csv(file_ngoaiq5, encoding='utf-8-sig')
df_vsa = pd.read_csv(file_vsa, encoding='utf-8-sig')
df_vsa_test = pd.read_csv(file_vsa_test, encoding='utf-8-sig')

print("=== THÔNG TIN FILE GỐC ===")
print(f"File ngoaiq5: {df_ngoaiq5.shape}")
print(f"Cột: {list(df_ngoaiq5.columns)}\n")

print(f"File vsa: {df_vsa.shape}")
print(f"Cột: {list(df_vsa.columns)}\n")

print(f"File vsa_test: {df_vsa_test.shape}")
print(f"Cột: {list(df_vsa_test.columns)}\n")

# Giữ chỉ cột review
df_ngoaiq5_cleaned = df_ngoaiq5[['review']].copy()
df_vsa_cleaned = df_vsa[['review']].copy()
df_vsa_test_cleaned = df_vsa_test[['review']].copy()

# Gộp 3 dataframe
df_merged = pd.concat([df_ngoaiq5_cleaned, df_vsa_cleaned, df_vsa_test_cleaned], ignore_index=True)

# Xóa review rỗng nếu có
df_merged['review'] = df_merged['review'].astype(str).str.strip()
df_merged = df_merged[df_merged['review'] != ''].reset_index(drop=True)

# Tạo review_id từ 00001, 00002, ...
df_merged.insert(0, 'review_id', df_merged.index + 1)
df_merged['review_id'] = df_merged['review_id'].apply(lambda x: f"{x:05d}")

print("=== FILE MERGED ===")
print(f"Kích thước sau merge: {df_merged.shape}")
print(f"Cột: {list(df_merged.columns)}\n")

# Lưu file
save_dir = '../dataset/preprocessed'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

output_path = os.path.join(save_dir, 'reviews_final_merged.csv')
df_merged.to_csv(output_path, index=False, encoding='utf-8-sig')

print(f"✅ Đã lưu file merged thành công tại: {output_path}")
print(f"📊 Tổng số dòng: {len(df_merged):,}")
print(f"📊 Số cột: {df_merged.shape[1]}")

print("\n--- 5 DÒNG ĐẦU ---")
display(df_merged.head())

print("\n--- 5 DÒNG CUỐI ---")
display(df_merged.tail())
```
