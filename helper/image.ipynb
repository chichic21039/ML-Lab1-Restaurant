```python
import img2pdf
from PIL import Image
import os

def png_to_pdf(input_path, output_path):
    print(f"--- Đang chuyển {input_path} sang PDF ---")
    # Đảm bảo file tồn tại
    if not os.path.exists(input_path):
        print("Lỗi: Không tìm thấy file đầu vào!")
        return

    # Chuyển đổi sang PDF
    with open(output_path, "wb") as f:
        f.write(img2pdf.convert(input_path))
    
    print(f"Hoàn thành! File PDF lưu tại: {output_path}")

# Sử dụng
png_to_pdf("robust.png", "robust.pdf")
```
