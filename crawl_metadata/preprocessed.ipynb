```python
import json

INPUT_FILE = "metadata.json"
OUTPUT_FILE = "metadata_preprocessed.json"

# Các khu vực hợp lệ
VALID_KEYWORDS = ["Chợ Lớn", "Chợ Quán", "An Đông"]

# Đọc dữ liệu
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

filtered_data = []

for place in data:
    avg_rating = place.get("avg_rating")
    address = place.get("address", "")

    # ❌ Bỏ nếu không có rating
    if avg_rating is None or avg_rating == "":
        continue

    # ❌ Bỏ nếu địa chỉ không chứa keyword
    if not any(keyword in address for keyword in VALID_KEYWORDS):
        continue

    # ✅ Giữ lại
    filtered_data.append(place)

# Ghi file output
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    json.dump(filtered_data, f, ensure_ascii=False, indent=2)

# Log kết quả
print(f"Tổng ban đầu: {len(data)}")
print(f"Sau khi lọc: {len(filtered_data)}")
print(f"Đã lưu vào: {OUTPUT_FILE}")
```

```python
import json

INPUT_FILE = "metadata_preprocessed.json"
OUTPUT_FILE = "url_quan_5_preprocessed.txt"

# Đọc dữ liệu
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    data = json.load(f)

# Lấy danh sách URL (field "url")
urls = []

for place in data:
    url = place.get("url")
    if url:  # tránh None
        urls.append(url)

# Ghi ra file txt (mỗi url 1 dòng)
with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
    for url in urls:
        f.write(url + "\n")

print(f"✔️ Đã lưu {len(urls)} URL vào {OUTPUT_FILE}")
```
